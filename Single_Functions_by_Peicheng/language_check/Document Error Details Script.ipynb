```python
import pandas as pd
import language_tool_python  as language_check
import docx
from datetime import date
from zipfile import ZipFile
#importing warnings
import warnings
warnings.filterwarnings("ignore")
import unidecode
import os
```

```python
Spelling = lambda x: True if (x.category == 'TYPOS') or (x.category == 'TYPOGRAPHY') else False

Punctuation = lambda x: True if x.category == 'PUNCTUATION' else False

Syntax = lambda x: True if x.category == 'STYLE' else False

Grammar= lambda x: True if (x.category == 'GRAMMAR') or (x.category == 'CASING') else False
```

```python
# Mention the language keyword
tool = language_check.LanguageTool("en")
```

```python
# Create a ZipFile Object and load sample.zip in it
with ZipFile('Total.zip', 'r') as zip:
   # Get list of files names in zip
   Files = zip.namelist()
```

```python
for i in range(len(Files)):
    Files[i]= Files[i].strip("Total")
```

## ERROR DETAILS

```python
#Defining Parent Directory
parent_dir = os.getcwd()
# Directory
directory = "Error_Details_2"
# Directory paths
path = os.path.join(parent_dir, directory)
# Create the directory
if not os.path.exists(path):
    os.mkdir(path)
else:
    pass
```

```python
df_errors_all= pd.DataFrame()
df_errors_all["File Name"]= ""
df_errors_all["Sentence"]= ""
df_errors_all["Error Message"]= ""
df_errors_all["replacements"]=""
df_errors_all["Character/ Start Position in Text"]=""
df_errors_all["Type"]= ""
df_errors_all["Subtype"]= ""
for j in range(len(Files)):
    try:
        doc = docx.Document("Total"+Files[j])
        all_paras = doc.paragraphs
        a=""
        Matches= ""
        for para in all_paras:
            a= a+"\n"+para.text
            a= a.strip("\n")
            a= unidecode.unidecode(a)
        Matches= tool.check(a)
        df_errors= pd.DataFrame()
        df_errors["File Name"]= [None]*len(Matches)
        df_errors["File Name"]= Files[j]
        df_errors["Sentence"]= ""
        df_errors["Error Message"]= ""
        df_errors["replacements"]=""
        df_errors["Character/ Start Position in Text"]=""
        df_errors["Type"]= ""
        df_errors["Subtype"]= ""
        for i in range(len(Matches)):
            df_errors["Sentence"][i]= Matches[i].sentence
            df_errors["Error Message"][i]= Matches[i].message
            df_errors["replacements"][i]= Matches[i].replacements
            df_errors["Character/ Start Position in Text"][i]= Matches[i].offset
            df_errors["Type"][i]= Matches[i].category
            df_errors["Subtype"][i]= Matches[i].ruleId
        df_errors_all= df_errors_all.append(df_errors,ignore_index=True)
        print(j)
    except:
        print("No Data: "+ str(j))
```

```python
df_errors_all.to_excel("Combined Error Details.xlsx", index=False)
```

```python
df= pd.DataFrame()
df["File_Name"] = [None]*len(Files)
df["Year"]= date.today().year
df["Month"]= date.today().month
df["Day"]= date.today().day
df["Spelling_Error"]=0
df["Grammar_Error"]=0
df["Punctuation_Error"]=0
df["Syntax_Error"]=0
for i in range(len(Files)):
    try:
        doc = docx.Document("Total"+Files[i])
        all_paras = doc.paragraphs
        a=""
        Matches= ""
        for para in all_paras:
            a= a+"\n"+para.text
            a= a.strip("\n")
            a= unidecode.unidecode(a)
        Matches= tool.check(a)
        final_spelling_errors=0
        final_grammar_errors=0
        final_syntax_errors=0
        final_punctuation_errors=0
        df["File_Name"][i] = Files[i]
        df["Year"][i]= datetime.utcfromtimestamp(os.stat("Total/"+Files[i]).st_mtime).strftime('%Y')
        df["Month"][i]= datetime.utcfromtimestamp(os.stat("Total/"+Files[i]).st_mtime).strftime('%m')
        df["Day"][i]= datetime.utcfromtimestamp(os.stat("Total/"+Files[i]).st_mtime).strftime('%d')
        for match in Matches:
            if Spelling(match)==True:
                final_spelling_errors=final_spelling_errors+1
            else:
                pass
        df["Spelling_Error"][i]= final_spelling_errors

        for match in Matches:
            if Grammar(match)==True:
                final_grammar_errors=final_grammar_errors+1
            else:
                pass
        df["Grammar_Error"][i]= final_grammar_errors

        for match in Matches:
            if Syntax(match)==True:
                final_syntax_errors=final_syntax_errors+1
            else:
                pass
        df["Syntax_Error"][i]= final_syntax_errors

        for match in Matches:
            if Punctuation(match)==True:
                final_punctuation_errors=final_punctuation_errors+1
            else:
                pass
        df["Punctuation_Error"][i]=final_punctuation_errors
        print(i)
    except:
        df["File_Name"][i] = Files[i]
        df["Year"][i]= datetime.utcfromtimestamp(os.stat("Total/"+Files[i]).st_mtime).strftime('%Y')
        df["Month"][i]= datetime.utcfromtimestamp(os.stat("Total/"+Files[i]).st_mtime).strftime('%m')
        df["Day"][i]= datetime.utcfromtimestamp(os.stat("Total/"+Files[i]).st_mtime).strftime('%d')
        df["Spelling_Error"][i]="Error while opening file"
        df["Grammar_Error"][i]="Error while opening file"
        df["Punctuation_Error"][i]="Error while opening file"
        df["Syntax_Error"][i]="Error while opening file"
        print(i)
```

### WORD COUNT

```python
OUT = 0
IN = 1
 
# Returns number of words in string
def countWords(string):
    state = OUT
    wc = 0
 
    # Scan all characters one by one
    for i in range(len(string)):
 
        # If next character is a separator,
        # set the state as OUT
        if (string[i] == ' ' or string[i] == '\n' or
            string[i] == '\t'):
            state = OUT
 
        # If next character is not a word
        # separator and state is OUT, then
        # set the state as IN and increment
        # word count
        elif state == OUT:
            state = IN
            wc += 1
 
    # Return the number of words
    return wc
```

```python
df["Word Count"]= 0
for i in range(len(df)):
    try:
        doc = docx.Document("Total"+Files[i])
        all_paras = doc.paragraphs
        a=""
        for para in all_paras:
            a= a+"\n"+para.text
            a= a.strip("\n")
            a= unidecode.unidecode(a)
        df["Word Count"][i]= countWords(a)
    except:
        print(i)
```
